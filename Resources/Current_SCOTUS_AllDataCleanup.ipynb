```python
import pandas as pd
import numpy as np
```

```python
csv_file = "SCDB_2018_02_justiceCentered_Citation.csv"
scotus = pd.read_csv(csv_file, encoding = "latin1")
scotus.head()
```

```python
current_df = scotus[["term", "justice", "direction", "issueArea", "caseOriginState", "precedentAlteration", "lawType"]]

current_df.head()
```

```python
current_df.count()
```

```python
noNA = current_df.dropna(axis=0, subset=["issueArea", "direction"])
```

```python
noNA.count()
```

```python
current_df = noNA.fillna(0)
current_df.count()
```

```python
current_df = current_df.astype({"term": int, "caseOriginState": int, "precedentAlteration": int, 
                                "issueArea": int, "lawType": int, 
                               "justice": int, "direction": int})
current_df.head()
```

```python
current_scotus = current_df[current_df["justice"] > 105]
current_scotus.head()
```

```python
current_scotus.to_csv("Current-SCOTUS-Data.csv")
```

```python
unique_term = current_scotus["term"].unique()
unique_issueArea = current_scotus["issueArea"].unique()
unique_caseOriginState = current_scotus["caseOriginState"].unique()
unique_lawType = current_scotus["lawType"].unique()

print(unique_term)
print(unique_lawType)
print(unique_issueArea)
print(unique_caseOriginState)
```

```python
possibilites = [unique_term.all(), 1953, 1960, 1962, 1965, 1968, 1972, 1973, 1977, 1984, 2018]
exists = current_scotus["term"].tolist()
difference = pd.Series([item for item in possibilites if item not in exists])
target = current_scotus["term"].append(pd.Series(difference))
target = target.reset_index(drop=True)
dummies_term = pd.get_dummies(target, prefix="term")
dummies_term = dummies_term.drop(dummies_term.index[list(range(len(dummies_term)-len(difference), len(dummies_term)))])
dummies_term = dummies_term.drop(["term_True"], axis=1)
dummies_term.head()
```

```python
possibilites = [unique_issueArea.all()]
exists = current_scotus["issueArea"].tolist()
difference = pd.Series([item for item in possibilites if item not in exists])
target = current_scotus["issueArea"].append(pd.Series(difference))
target = target.reset_index(drop=True)
dummies_issueArea = pd.get_dummies(target, prefix="issueArea")
dummies_issueArea = dummies_issueArea.drop(dummies_issueArea.index[list(range(len(dummies_issueArea)-len(difference), len(dummies_issueArea)))])
dummies_issueArea.head()
```

```python
possibilites = [unique_caseOriginState.all(), 60]
exists = current_scotus["caseOriginState"].tolist()
difference = pd.Series([item for item in possibilites if item not in exists])
target = current_scotus["caseOriginState"].append(pd.Series(difference))
target = target.reset_index(drop=True)
dummies_caseOriginState = pd.get_dummies(target, prefix="caseOriginState")
dummies_caseOriginState = dummies_caseOriginState.drop(dummies_caseOriginState.index[list(range(len(dummies_caseOriginState)-len(difference), len(dummies_caseOriginState)))])
dummies_caseOriginState.head()
```

```python
possibilites = [unique_lawType.all()]
exists = current_scotus["lawType"].tolist()
difference = pd.Series([item for item in possibilites if item not in exists])
target = current_scotus["lawType"].append(pd.Series(difference))
target = target.reset_index(drop=True)
dummies_lawType = pd.get_dummies(target, prefix="lawType")
dummies_lawType = dummies_issueArea.drop(dummies_lawType.index[list(range(len(dummies_lawType)-len(difference), len(dummies_lawType)))])
dummies_lawType.head()
```

```python
dummies_precAlt = pd.get_dummies(current_scotus["precedentAlteration"], prefix="precedentAlteration")
dummies_precAlt.reset_index(drop=True, inplace=True)
dummies_precAlt.head()
```

```python
dir_df = pd.DataFrame(current_scotus["direction"].reset_index(drop=True))
justice_df = pd.DataFrame(current_scotus["justice"].reset_index(drop=True))

OH_scotus = pd.concat([dummies_term, dummies_issueArea], axis=1)
OH_scotus = pd.concat([OH_scotus, dummies_caseOriginState], axis=1)
OH_scotus = pd.concat([OH_scotus, dummies_lawType], axis=1)
OH_scotus = pd.concat([OH_scotus, dummies_precAlt], axis=1)
OH_scotus = pd.concat([OH_scotus, dir_df], axis=1)
OH_scotus = pd.concat([OH_scotus, justice_df], axis=1)
OH_scotus.head()
```

```python
OH_scotus.to_csv("SCOTUS_OHEncoded.csv")
```
