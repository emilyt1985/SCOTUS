```python
import pandas as pd
import numpy as np
```

```python
csv_file = "SCDB_2018_02_justiceCentered_Citation.csv"
csv_file2 = "appointment_party.csv"
```

```python
scotus = pd.read_csv(csv_file, encoding = "latin1")
scotus.head()
```

```python
vote_df = scotus[["justice", "justiceName", "majVotes", "minVotes", "decisionDirection", "majority", "caseId", "term", "partyWinning", "precedentAlteration", "vote", "issueArea"]]

vote_df.head()
```

```python
votes = vote_df.fillna(0)
```

```python
votes["vote_direction"]= 0
```

```python
vote = votes.astype({"partyWinning": int, "precedentAlteration": int, "majority": int, "vote_direction": int, "decisionDirection": int, "vote": int, "issueArea": int})
vote.dtypes
```

```python
first = vote.query("majority == 1 & decisionDirection == 2").index
second = vote.query("majority == 1 & decisionDirection == 1").index
third = vote.query("majority == 2 & decisionDirection == 1").index
fourth = vote.query("majority == 2 & decisionDirection == 2").index
fifth = vote.query("decisionDirection == 3").index
```

```python
vote.loc[first, "vote_direction"] = 1
vote.loc[second, "vote_direction"] = 2
vote.loc[third, "vote_direction"] = 1
vote.loc[fourth, "vote_direction"] = 2
vote.loc[fifth, "vote_direction"] = 3
vote.head(25)
```

```python
vote.to_csv("full_votes.csv")
```
