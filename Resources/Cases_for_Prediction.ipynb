```python
import pandas as pd
import numpy as np
```

```python
#  Read CSVs
file1 = "SCDB_2018_02_caseCentered_Citation.csv"
file2 = "landmark.csv"
file3 = "currentcases_cleaned.csv"

SC_data = pd.read_csv(file1, encoding = "latin1")
LM_data = pd.read_csv(file2)
CC_data = pd.read_csv(file3)

# LM_data
# SC_data.head()
CC_data.head()
```

```python
# Select Landmark cases from Supreme Court data set
caseIds = LM_data["CaseId"]
# caseIds
LM_SC = SC_data[SC_data["caseId"].isin(caseIds)]
LM_SC.head()
```

```python
LM_SC_data = LM_SC[["caseId","caseName", "term", "precedentAlteration", "issueArea", "lawType", "caseOriginState", "decisionDirection"]]
LM_SC_data.head()
```

```python
# replace NaN with 9 (no legal provision) for lawType and 60 (United States) for caseOriginState
LM_SC_data = LM_SC_data.fillna(value={"lawType": 9, "caseOriginState": 60})
LM_SC_data
```

```python
CC_data = CC_data[["caseId","caseName", "term", "precedentAlteration", "issueArea", "lawType", "caseOriginState", "decisionDirection"]]
CC_data["caseName"] = CC_data['caseName'].str.upper()
CC_data
```

```python
cases_df = pd.concat([LM_SC_data, CC_data], axis=0)
cases_df.reset_index(drop=True, inplace=True)
cases_df
```

```python
cases_df.to_csv("cases_for_prediction.csv")
```

```python
# grab unique values to include in one-hot encoding later
unique_term = cases_df["term"].unique()
unique_issueArea = cases_df["issueArea"].unique()
unique_caseOriginState = cases_df["caseOriginState"].unique()
unique_lawType = cases_df["lawType"].unique()

print(unique_term)
print(unique_lawType)
print(unique_issueArea)
print(unique_caseOriginState)
```
