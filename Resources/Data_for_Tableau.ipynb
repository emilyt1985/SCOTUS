```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# Set file path
csv_file = "SCDB_2018_02_justiceCentered_Citation.csv"
```

```python
# import csv
scotus = pd.read_csv(csv_file, encoding = "latin1")
scotus.head()
```

```python
# Set columns
vote_df = scotus[["justice", "justiceName", "majVotes", "direction", "minVotes", "decisionDirection", "majority", "caseId", "term", "partyWinning", "precedentAlteration", "vote", "issueArea"]]

vote_df.head()
```

```python
# Drop some NAs
votes = vote_df.dropna(axis=0, subset=['issueArea'])
```

```python
# Set others to zero
vote_counts = votes.fillna(0)
```

```python
# Set floats to integers
vote = vote_counts.astype({"partyWinning": int, "direction": int, "precedentAlteration": int, "majority": int, "decisionDirection": int, "vote": int, "issueArea": int})
```

```python
# Export to csv
vote.to_csv("full_votes.csv")
```
