```python
import pandas as pd
import numpy as np
```

```python
csv_file = "SCDB_2018_02_justiceCentered_Citation.csv"
scotus = pd.read_csv(csv_file, encoding = "latin1")
scotus.head()
```

```python
current_df = scotus[["lexisCite", "term", "justice", "justiceName", "direction", "majority", "majVotes", "minVotes",
                      "issueArea", "decisionDirection", "caseOriginState", "precedentAlteration", "lawType"]]

current_df.head()
```

```python
noNA = current_df.dropna(axis=0)
```

```python
current_df = noNA.fillna(0)
current_df.dtypes
```

```python
current_df = current_df.astype({"term": int, "caseOriginState": int, "precedentAlteration": int, 
                                "issueArea": int, "decisionDirection": int, "lawType": int, "majVotes": int,
                               "minVotes": int, "majority": int, "justice": int, "direction": int})
current_df.head()
```

```python
current_scotus = current_df[current_df["justice"] > 107]
current_scotus.head()
```

```python
current_scotus.to_csv("Current-SCOTUS-Data.csv")
```
